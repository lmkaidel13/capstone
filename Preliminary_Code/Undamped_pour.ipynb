```python
from matplotlib import pyplot
%matplotlib inline

import numpy as np

import pandas as pd
```

```python
CFD = pd.read_csv("CFD_raw_tester.csv",delimiter=',', names=['iteration','continuity','x_v','y_v','vf','t'])
CFD=CFD[1:]
```

```python
for i in range(len(CFD['t'])):
    temp=CFD['t'][i+1].split(':')
    CFD['t'][i+1]=float(temp[-1])
```

```python
for i in range(len(CFD['t'])-1):
    CFD['t'][i+2]=CFD['t'][i+2]+CFD['t'][i+1]
```

```python

pyplot.yticks([],"")

pyplot.plot(CFD['t'],CFD['x_v'],label='x')
pyplot.plot(CFD['t'],CFD['y_v'],label='y')
pyplot.xlabel('time [s]')
pyplot.ylabel('velocity')
pyplot.legend(loc='best');
```
